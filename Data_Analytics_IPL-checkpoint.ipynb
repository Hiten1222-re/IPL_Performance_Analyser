```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.expand_frame_repr', False)

a=pd.read_excel('D:\deliveries.xlsx')
print(a.head())
```

```python
print(a['batting_team'].unique())
```

```python
# This shows the total score in the match by teams
team= a.groupby(["match_id", "batting_team"]).agg({'total_runs':'sum','is_wicket':'sum','over':'max'}).reset_index().head()

print(team)
```

```python
# This calculates the number of wickets taken by a bowler
bowler=a.groupby(['match_id','bowling_team','bowler'])['is_wicket'].sum().reset_index().head(10)
print(bowler)
```

```python
# This calculates the number of wickets taken by a bowler in powerplay, middle and death overs
def phasee(over):
    if over <=6:
        return 'Powerplay'
    elif over >16:
        return 'Death'
    else:
        return 'Middle'

a['phase']=a['over'].apply(phasee)

phase_wise_wickets= a.groupby(['match_id','bowling_team','bowler','phase'])['is_wicket'].sum()
phase_wise_wickets=phase_wise_wickets[phase_wise_wickets>0]
print(phase_wise_wickets.head(20))
```

```python
# Top run-scorers from all phases
performance_during_phases=a.groupby(['match_id','batting_team','batter','phase'])['total_runs'].sum().reset_index()
print(performance_during_phases.sort_values(['phase','total_runs'],ascending=False).head(5),end='\n')
print(performance_during_phases[performance_during_phases['phase']!='Powerplay'].sort_values(['phase','total_runs'],ascending=False).head(5),end='\n')
print(performance_during_phases[performance_during_phases['phase']=='Death'].sort_values(['phase','total_runs'],ascending=False).head(5),end='\n')
```

```python
# Top Batsman in phases
con=a.groupby(['match_id','phase','batter'],as_index=False)['total_runs'].sum()
tot=con.sort_values('total_runs',ascending=False).drop_duplicates('phase')
print(tot)
```

```python
# Best Batsman phasewise
con=a.groupby(['phase','batter'],as_index=False)['total_runs'].sum()
tot=con.sort_values('total_runs',ascending=False).drop_duplicates('phase')
print(tot)
```

```python
# Virat Kohli score comparison between phases
vir=a.groupby(['phase','batter'],as_index=False)['total_runs'].sum()
virat=vir[vir['batter']=='V Kohli']
print(virat)
prun=int(virat[virat['phase']=='Powerplay']['total_runs'])
mrun=int(virat[virat['phase']=='Middle']['total_runs'])
cal=(mrun-prun)*100/prun
# print(cal)
print(f'Virat Kohli plays better in middle overs by {cal:.2f}% compared to powerplay')
```

```python
# Line plot of Season-wise average estimated scores of Royal Challengers Bangalore
team_runs=a.groupby(['match_id','batting_team'],as_index=False)['total_runs'].sum()
team_runs=team_runs[team_runs['batting_team']=='Royal Challengers Bangalore']
avg=[]
while len(team_runs) > 0:
    con1=team_runs.head(20)
    con2=con1['total_runs'].mean()
    avg.append(con2)
    team_runs=team_runs.iloc[20:].reset_index(drop=True)
season=list(range(1,len(avg)+1))
plt.plot(season,avg)
plt.show()
```

```python
# Improvement of best team over the seasons
teams_score_list=a.groupby(['batting_team'],as_index=False)['total_runs'].sum()
best=teams_score_list[teams_score_list['total_runs']==teams_score_list['total_runs'].max()]
teamlist=a.groupby(['match_id','batting_team'],as_index=False)['total_runs'].sum()
best_team_results=teamlist[teamlist['batting_team']==best['batting_team'].iloc[0]]
avg=[]
while len(best_team_results) > 1:
    con1=best_team_results.head(18)
    con2=con1['total_runs'].mean()
    avg.append(con2)
    best_team_results=best_team_results.iloc[19:].reset_index(drop=True)

season=list(range(1,len(avg)+1))
plt.plot(season,avg)
```
